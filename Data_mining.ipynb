<h2 style="color:#7763A8;" > Pr√©traitement des donn√©es </h2>
<p> Cette partie consiste a nettoyer les donn√©es , Enlever les @ , les symboles ,  les emojis , les flags </p>

```python
import re
for index, row in combined_csv.iterrows():
    err = row['tweetText']
    new0 = re.sub(r"http\S+", "", err)
    new1 = re.sub(r"#\S+", "", new0)
    new2 = re.sub(r"@\S+", "", new1)
    new3 = re.sub(r"\n+", "", new2)
    new4 = re.sub(r"RT+", "", new3)
    new5 = re.sub("hhh+", '', new4)
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    new6 = re.sub(emoji_pattern, "", new5)
    combined_csv.loc[index,'tweetText'] =new6
    
```


<h3 style="color:#7763A8;"> Tokenize mots et phrases avec NLTK</h3>

<p> La tokenisation est le processus par lequel une grande quantit√© de texte est divis√©e en parties plus petites appel√©es jetons .

Le traitement du langage naturel est utilis√© pour cr√©er des applications telles que la classification de texte, le chatbot intelligent, l'analyse sentimentale, la traduction de la langue, etc. Il devient vital de comprendre le mod√®le dans le texte pour atteindre l'objectif susmentionn√©. Ces jetons sont tr√®s utiles pour trouver de tels mod√®les et sont consid√©r√©s comme une √©tape de base pour la racine et la lemmatisation </p> 

<p> Le Code suivant permet d'√©liminer ces mots de notre texte  : </p>

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
ps = PorterStemmer() 
stemed_dataset=[]
for i in range(0,tweeter.shape[0]):
    stemmed_array=tweeter['tweetText'][i].split()
    stemmed=[ps.stem(word) for word in stemmed_array if not word in set(stopwords.words('english'))]
    stemmed=' '.join(stemmed)
    stemed_dataset.append(stemmed)
print(stemed_dataset[0:10])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X=cv.fit_transform(stemed_dataset)
print(X)
```

<h2 style="color:#7763A8;"> Classification des tweets </h2>

<p> Clustering K-Means. </p>
<p> Le nombre de clusters √† former ainsi que le nombre de centres de gravit√© √† g√©n√©rer.</p>

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,30):
    Kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0,verbose=True)
    Kmeans.fit(X)
wcss.append(Kmeans.inertia_)
```

```python
true_k=30
Kmeans=KMeans(n_clusters=true_k,init='k-means++',n_init=1)
Kmeans.fit(X)
```

<p> Apr√©s fixer le nombre de clusters, le model KMeans va nous donner une id√©e g√©n√©rale sur les cluster avec une disaine de mots   puis afficher la liste des 30 clusters avec un echantillon representant les tweets de chaque cluster. </p>

```python

print("Top terms per cluster:")
order_centroids = Kmeans.cluster_centers_.argsort()[:, ::-1]
terms = cv.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i)
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind])
    print()
print("\n")
```

```python

i=0
j=0
while i<30:
    while True: 
        Y=cv.transform([stemed_dataset[j]])
        prediction=Kmeans.predict(Y)
        if i == prediction:
            print("tweet of cluster "+str(prediction)+tweeter.loc[j,"tweetText"]) 
            j=0
            break
        j+=1
    i+=1
```
